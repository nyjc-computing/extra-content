# Formatted text (pre-2007)

Before 2007, the primary way to get formatted text was using HTML.

People had to learn to read and understand HTML tags to imagine what the text looked like.

This was highly inconvenient for developers, who worked primarily in text editors or IDEs that do not interpret HTML.

# The rise of Markdown (2004 onwards)

In 2004, John Gruber, a developer, came up with an idea: Why not use symbols to make text more readable, *and* also represent text formatting?

This way of writing formatted text without tags came to be known as [Markdown](https://daringfireball.net/projects/markdown/).

In fact, Jupyter Notebooks use Markdown for formatting in Markdown cells! You can edit the Markdown cells to see how Markdown formatting is applied.

# Using Markdown

Main article: [Mastering Markdown](https://guides.github.com/features/mastering-markdown/#syntax)

See this article for the full Markdown syntax guide.

A Markdown document is readable both in plain text, as well as when viewed as a formatted HTML document.

## Basic formatting

### Bold

Markdown uses double asterisks (`**`) or double underscores (`__`) for bold.

**Markdown:**
```
**This will be bolded.**

__This will be bolded too.__
```

**Result:**

**This will be bolded.**

__This will be bolded too.__

### Italics

Markdown uses single asterisk (`*`) or single underscore (`_`) for italics.

**Markdown:**
```
*This will be italic.*

_This will be italic too._
```

**Result:**

*This will be italic.*

_This will be italic too._

### Multiline text

If you write text on multiple lines, Markdown treats it as part of the same text block.

**Markdown:**

```
Writing text
on multiple
lines
does not work
```

**Result:**

Writing text
on multiple
lines
does not work

To force Markdown to recognise the line breaks, put two spaces after each line.

**Markdown:**

```
Put two spaces  
after each line  
to make Markdown  
recognise line breaks
```

**Result:**

Put two spaces  
after each line  
to make Markdown  
recognise line breaks



## Headers

Markdown uses the asterisk (`#`) to indicate header levels.

**Markdown:**

`# This is header 1`

**Result:**

# This is header 1

**Markdown:**

`### This is header 3`

**Result:*8

### This is header 3

## Lists

To create a list, add list elements with a hyphen (`-`).

**Markdown:**
```
This is a list
- list item 1
- list item 2
```

**Result:**

This is a list
- list item 1
- list item 2

For an ordered list, use numbers instead of hyphens:

**Markdown:**
```
This is a list
1. list item 1
2. list item 2
```

**Result:**

This is a list
1. list item 1
2. list item 2
